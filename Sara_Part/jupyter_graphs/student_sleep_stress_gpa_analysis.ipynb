```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("proj/cleaned_student_dataset.csv") 
```

```python
cols = [
    "sleep_hours_per_day",
    "study_hours_per_day",
    "extracurricular_hours_per_day",
    "social_hours_per_day",
    "physical_activity_hours_per_day",
    "gpa",
    "stress_level",
    "total_engagement_hours"
]

df_corr = df[cols].corr()
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(df_corr, annot=True, cmap="coolwarm", vmin=-1, vmax=1, linewidths=.5)
plt.title("Correlation Heatmap of Student Lifestyle Variables")
plt.show()
```

```python
# For this heatmap, I wanted to look at how the main lifestyle variables in the student dataset relate to each other at the same time.
# Each square shows the exact correlation value, and the colors help me quickly see whether the relationship is positive or negative.
# From the chart, I can see that study_hours_per_day has a strong positive relationship with gpa, which fits the idea that students
# who study more usually earn higher grades. Study hours are also highly related to stress_level, so students who spend more time
# studying tend to report feeling more stressed. Sleep_hours_per_day has a moderate negative correlation with stress_level and
# total_engagement_hours, which suggests that students who are busier or more overwhelmed are sleeping less. Physical_activity_hours_per_day
# also shows negative relationships with sleep and study time, which might mean that students who exercise more have less time left
# for academics or rest. Overall, this heatmap gives me a clear summary of which habits are most closely linked and helps explain why
# patterns like higher workload, higher stress, and lower sleep appear so strongly in the rest of my analysis.
```
